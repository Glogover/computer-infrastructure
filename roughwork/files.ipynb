# File Listings

https://docs.python.org/3/library/os.html

```python
# Listing files in a folder.
import os

# Data frames.
import pandas as pd
```

```python
# Try it out
data_files = os.listdir('../data/')

# Show
data_files
```

```python
# Sort the list of files.

sorted(data_files, reverse=True)

# Not modified.

data_files
```

```python
# Sort the lst of files
data_files.sort(reverse=True)

# Show 
data_files
```

```python
# The latest file.
data_files[0]
```
