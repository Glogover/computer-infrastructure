```python
# Current date and time.
now = dt.datetime.now()

# Show
now
```

```python
# Format date and time.
now.strftime("%Y-%m-%d %H:%M:%S") # Format as "YYYY-MM-DD HH:MM:SS".
```

```python
# Format date and time.
now.strftime("%Y%m%d-%H%M%S") # Format as "YYYYMMDD-HHMMSS".
```

```python
# File name.
"../data/" + now.strftime("%Y%m%d-%H%M%S") + ".csv" 
```

```python
# Save data as CSV.
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html

df.to_csv("../data/" + dt.datetime.now().strftime("%Y%m%d-%H%M%S") + ".csv") # Save with timestamp in filename.
```
